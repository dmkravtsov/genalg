```python
import optuna
import numpy as np
import time

#   Rastrigin
def rastrigin(trial):
    #   
    dim = 10
    #   
    x = [trial.suggest_float(f"x{i}", -5.12, 5.12) for i in range(dim)]
    
    #    Rastrigin
    A = 10
    return A * dim + sum([(xi ** 2 - A * np.cos(2 * np.pi * xi)) for xi in x])

#      
results = []

#  
def optimize_rastrigin():
    start_time = time.time()  #   
    study = optuna.create_study(direction="minimize", sampler=optuna.samplers.TPESampler(seed=42))
    
    #        
    def callback(study, trial):
        #    
        results.append({
            'trial_number': trial.number,
            'value': trial.value,
            'params': trial.params,
            'time': time.time() - start_time  #     
        })

    study.optimize(rastrigin, n_trials=1000, callbacks=[callback])  # 500 

    #   
    best_trial = study.best_trial
    best_result = None
    for result in results:
        if result['trial_number'] == best_trial.number:
            best_result = result
            break

    #  
    print(f" : {best_result['trial_number']}")
    print(f"   : {best_result['time']:.2f} ")
    print(f"  : {best_result['value']}")
    print(f"  : {best_result['params']}")
    print(f" : {len(results)}")

#  
optimize_rastrigin()
```
